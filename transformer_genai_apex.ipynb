```python
from transformers import pipeline
```

```python
import pandas as pd
```

```python
data=pd.read_html("https://en.wikipedia.org/wiki/National_Institutes_of_Technology_(India)",header=0)[2]
```

```python
data=data.astype(str)
```

```python
from transformers import AutoTokenizer,AutoModelForTableQuestionAnswering
```

```python
model="google/tapas-large-finetuned-wtq"
```

```python
tokenizer=AutoTokenizer.from_pretrained('google/tapas-large-finetuned-wtq')
```

```python
tokenizer
```

```python
model=AutoModelForTableQuestionAnswering.from_pretrained(model)
```

```python
tqapipe=pipeline(task="table-question-answering",model=model,tokenizer=tokenizer)
```

```python
data
```

```python
tqapipe(table=data,query="Name the NIT founded prior to 1965")
```
